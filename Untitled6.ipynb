```python
import datetime as dt
import yfinance as yf
import pandas as pd


tikers = ["AMZN","AAPL","MSFT","GOOG"]

start = dt.datetime.today()-dt.timedelta(3650)
end = dt.datetime.today()
cl_price = pd.DataFrame()   # data frame filed with closing price

for stock in tikers:
    cl_price[stock] = yf.download(stock,start,end)["Close"]

cl_price.dropna(axis=0,how ="any",inplace=True)

daily_returns = cl_price.pct_change()  #dataframe for each stok created.
cl_price.plot(subplots=True,layout=(2,2),title="Stock Price",grid = True)

# daily_returns.plot(subplots=True, layout=(2,2), title="Daily Returns", grid = True)
(1+daily_returns).cumprod().plot(grid = True)

# cl_price.plot(subplots=True,layout=(2,2),title="Stock Price",grid = True)




```
